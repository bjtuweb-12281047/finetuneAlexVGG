```python
import numpy as np
import tensorflow as tf

from matplotlib import pyplot as plt
from preprocessing import bgr_crop_prep, bgr_resize_prep, crop_prep, resize_prep, vgg_prep, inception_prep
```

```python
img_size      = 224
img_path      = './images/llama.jpeg'
img_file      = tf.read_file(img_path)
img_decoded   = tf.image.decode_jpeg(img_file, channels=3)

bgr_crop = bgr_crop_prep.preprocess_image(img_decoded, img_size, img_size)
bgr_resize = bgr_resize_prep.preprocess_image(img_decoded, img_size, img_size)

crop = crop_prep.preprocess_image(img_decoded, img_size, img_size)
resize = resize_prep.preprocess_image(img_decoded, img_size, img_size)

vgg = vgg_prep.preprocess_image(img_decoded, img_size, img_size)
inception = inception_prep.preprocess_image(img_decoded, img_size, img_size)
```

```python
with tf.Session() as sess:
    img = sess.run(img_decoded)
    bgg_crop = sess.run(bgr_crop)
    bgr_resize = sess.run(bgr_resize)
    crop = sess.run(crop)
    resize = sess.run(resize)
    vgg = sess.run(vgg)
    inception = sess.run(inception)
```

```python
    plt.figure()
    plt.imshow(img.astype(np.uint8))
    plt.suptitle("Loaded image")
    plt.show()
    
    plt.figure()
    plt.suptitle("vgg/inception")
    plt.subplot(121)
    plt.imshow(vgg.astype(np.uint8))
    plt.subplot(122)
    plt.imshow(inception.astype(np.uint8))
    plt.show()
    
    plt.figure()
    plt.suptitle("crop/resize")
    plt.subplot(121)
    plt.imshow(crop.astype(np.uint8))
    plt.subplot(122)
    plt.imshow(resize.astype(np.uint8))
    plt.show()
    
    plt.figure()
    plt.suptitle("BGR crop/resize:")
    plt.subplot(121)
    plt.imshow(brg_crop.astype(np.uint8))
    plt.subplot(122)
    plt.imshow(bgr_resize.astype(np.uint8))
    plt.show()    
```
